## Exercice 3

On nous informe que les objets proviennent de deux entrepôts différents, identifiés par le préfixe de leur `id` :

AM-... : Entrepôt d'Amsterdam.

BER-... : Entrepôt de Berlin.

La partie alphanumérique qui suit est le véritable identifiant des objets.

Nous souhaitons nettoyer les identifiants afin de pouvoir trier les objets en fonction de leur véritables identifiants. Il s'agit donc de renvoyer l'information de l'entrepôt d'origine à une autre colonne.

1. Créez une nouvelle colonne `entrepot`. Remplissez-la avec la valeur "Amsterdam" si l'id commence par "AM-", et "Berlin" s'il commence par "BER-".

Indice :  Vous pouvez utiliser la fonction `np.where` (**[documentation](https://numpy.org/devdocs/reference/generated/numpy.where.html)**).

```python
df['entropot']=np.where(df['id'].str.startswith('AM-'),'Amsterdam','Berlin')
```

2. Quel est le prix moyen des objets venant d'Amsterdam ? Celui des objets venant de Berlin ?

```python
print(df.groupby('entropot')['prix'].mean())
```

3. Modifiez la colonne `id` pour ne conserver que le code alphanumérique (la partie après le tiret -).

```python
df['id']=df['id'].str.split("-",n=1).str[1]
```

4. Triez le DataFrame par ordre alphabétique des nouveaux id, puis réinitialisez l'index pour qu'il soit continu (de 0 à N-1).

```python
df.sort_values(by='id').reset_index(drop=True)
```

## Exercise 5

Nous avons un DataFrame `pixels` de la bonne dimension mais il contient des éléments de type `str`. Nous devons donc convertir ce DataFrame au format numérique. De plus, nous savons que ce jeu de donnée est corrompu. Il nous faut donc vérifier que chaque pixel est bien un entier entre 0 et 255. Il faudra alors supprimer toutes les lignes avec des pixels corrompus. Il faudra faire attention à bien retenir quelles lignes sont supprimées car il faudra également enlever ces lignes du DataFrame `prix`.

1. Utilisez `pd.to_numeric` (**[documentation](https://pandas.pydata.org/docs/reference/api/pandas.to_numeric.html)**)pour convertir `pixels` au format numérique en faisant en sorte que tout ce qui n'est pas un nombre soit converti en `NaN`.

```python
pixels=pixels.apply(pd.to_numeric,errors='coerce')
```

2. Combien de cellules (pixels) contenaient autre chose qu'un nombre ?

```python
nb_erreurs = pixels.isna().sum().sum()
print(nb_erreurs)
```

3. Vérifiez si tous les cellules ayant un élément numérique contient bien la représentation mathématique d'un pixel (un entier entre 0 et 255) en affichant les minimum et maximum que l'on trouve dans `pixels`.

```python
maximum = pixels.max().max()
minimum = pixels.min().min()

print(f"minimum : {minimum}") 
print(f"max : {maximum}")
```

4. Créez un masque booléen `a_jeter` couvrant les lignes avec des valeurs corompues (`Nan` ou nombres absurdes). Cela représente combien de lignes ? 

```python
lignes_nan = pixels.isna().any(axis=1)
lignes_abs= ((pixels<0) | (pixels>255)).any(axis=1)
a_jeter = lignes_nan | lignes_abs
print(a_jeter.sum())
```

5. Enlevez de `pixels` et de `prix` les lignes couverte par `a_jeter`. Affichez les nouvelles dimensions.

```python
pixels= pixels.loc[~a_jeter].copy()
prix= prix.loc[~a_jeter].copy()
print(pixels.shape)
print(prix.shape)
```
